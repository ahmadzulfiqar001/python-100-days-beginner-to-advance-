```python
# import libraries

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# load dataset

kashti = sns.load_dataset("titanic")
kashti

#draw a plot

sns.barplot(x="class", y="fare", hue="sex", data=kashti, estimator= np.mean)
plt.show()
```
