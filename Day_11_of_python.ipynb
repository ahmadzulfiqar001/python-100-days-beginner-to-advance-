```python
# import libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#load dataset
kashti = sns.load_dataset("titanic")
sns.boxplot(x="survived",
            y="age", showmeans=True,
            meanprops={"marker":"+",
                       "markersize":"12",
                       "markeredgecolor":"red"},
            data=kashti )

plt.show()
```
