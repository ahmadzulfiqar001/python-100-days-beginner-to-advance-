```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")

sns.catplot(
    x="survived",
    y="age",
    col="sex",
    kind="box",
    showmeans=True,
    meanprops={
        "marker":"+",
        "markersize":"12",
        "markeredgecolor":"red"
    },
    data=kashti
)

# sns.catplot(
#     x="survived",
#     y="age",
#     col="sex",              # 🔥 facet here
#     kind="box",
#     showmeans=True,
#     meanprops={
#         "marker": "+",
#         "markersize": 12,
#         "markeredgecolor": "red"
#     },
#     data=kashti
# )

plt.show()
```

```python
sns.catplot(
    x="survived",
    y="age",
    col="class",        # 1st, 2nd, 3rd
    col_wrap=2,         # wrap like facet_wrap
    kind="box",
    showmeans=True,
    meanprops={
        "marker": "+",
        "markersize": 12,
        "markeredgecolor": "red"
    },
    data=kashti
)
```
