In Python, an identifier is a user-defined name used to identify a variable, function, class, module, or other object. It acts as a label that allows you to refer to data or logic throughout your program


Rules for Naming Identifiers
To be valid, an identifier must follow these strict rules:
Character Set: Can only contain letters (A-Z, a-z), digits (0-9), and underscores (_).
Start Character: Must start with a letter or an underscore; it cannot start with a digit.
No Keywords: Reserved Python Keywords (like if, while, class) cannot be used as identifiers.
No Special Characters: Symbols like @, $, #, and % are not allowed.
Case-Sensitivity: Python is case-sensitive, so myVar, MyVar, and MYVAR are three distinct identifiers.
