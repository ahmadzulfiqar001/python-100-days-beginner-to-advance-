**Type Conversion**

```python
x = 4 #integer
y = 4.0 #float
z = "Hello" #String

# x = x*y
# y= x*y

x=x+y

print(type(x))
print(type(y))
print(type(z))
```

```python
# implicit type conversion

x = 4 #integer
y = 4.0 #float
z = "Hello" #String

x=x+y

print(x, "type of x is:" , type(x))
print(type(y))
print(type(z))
```

```python
# explicit type conversion

age = input("What is your age?")
print(age , type(age))
#print(age , type(int(age)))
print(age , type(float(age)))


name = input("What is your name?")
print(name, type(name))
```
