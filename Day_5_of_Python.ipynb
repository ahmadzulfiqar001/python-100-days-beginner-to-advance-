**Loops**

```python
# important topic

# 2 types
# while loops
# for loops
```

```python
# while loops

x = 0
while x <= 10:
  print(x)

  x = x+1
```

```python
# for loop

for x in range (4,11):
  print(x)
```

```python
# array

days = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

for d in days:
  print(d)
```

```python
# array

days = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

for d in days:
  #if (d == "Wednesday"):    #loops stops
    # break
  if (d == "Wednesday"):
    continue            # skip d in this
  print(d)
```
